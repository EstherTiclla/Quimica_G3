```python
df = pd.read_csv("LeyDeLussac.csv")
df
```

```python
#Regresion Lineal Ley de Gussac
componentes = np.polyfit(df['x'], df['y'], 1)
linea_tendencia = np.poly1d(componentes)

# Crea el gráfico de los datos originales
plt.plot(df['x'], df['y'], label='Puntos de dispersion', marker='o', linestyle='None', color="blue")

# Crea el gráfico de la línea de tendencia
# Linspace para vectores equidistantes
x_tendencia = np.linspace(df['x'].min(), df['x'].max(), 100)
y_tendencia = linea_tendencia(x_tendencia)
plt.plot(x_tendencia, y_tendencia, label='Línea de Tendencia', color='red', linewidth=2.)

# Establece etiquetas de ejes y título
plt.xlabel('Tº Kelvin')
plt.ylabel('Volumen en ml')
plt.title('Gráfica Ley de Gussac')

# Cuadrícula
plt.grid(True)
plt.legend()

plt.show()
```

```python
df2 = pd.read_csv("LeyDeBoyle.csv")
df2
```

```python
df2['x_inversa'] = 1 / df2['x']
```

```python
df2
```

```python
#polifit devuelve la pendiente y la interseccion para formar la ecuacion
componentes = np.polyfit(df2['x_inversa'], df2['y'], 1)
linea_tendencia_inversa = np.poly1d(componentes)

# Crea el gráfico de la inversa de los datos
plt.plot(df2['x_inversa'], df2["y"], label='Datos', marker='o', linestyle='None', color='green')

# Crea el gráfico de la línea de tendencia de la inversa
x_tendencia_inversa = np.linspace(df2['x_inversa'].min(), df2['x_inversa'].max(), 100)
y_tendencia_inversa = linea_tendencia_inversa(x_tendencia_inversa)
plt.plot(x_tendencia_inversa, y_tendencia_inversa, label='Regresión lineal', color='blue', linewidth=4.0)

# Establece etiquetas de ejes y título
plt.xlabel('Volumen (litros)')
plt.ylabel('Presión (atm)')
plt.title('Gráfica Ley de Boyle')

# Agrega una cuadrícula y una leyenda
plt.grid(True)
plt.legend()

# Muestra el gráfico
plt.show()
```
