```python
from chempy import balance_stoichiometry

print("\nLa ecuacion del grafico es: C4H8 + O2 --> CO2 + OH2\n")

#Ingresamos los subindices

print("La ecuacion es: C[]H[] + O[] --> C[]O[] + O[]H[]\n")
print("\nIngrese los subindices para C[]H[] ")
sub_ind_C_r = input("C[] -->")
sub_ind_H_r = input("H[] -->")
print("\nIngrese los subindices para O[] ")
sub_ind_O_r = input("O[] -->")
print("\nIngrese los subindices para C[]O[] ")
sub_ind_C_p = input("C[] -->")
sub_ind_O_p1 = input("O[] -->")
print("\nIngrese los subindices para O[]H[] ")
sub_ind_O_p2 = input("O[] -->")
sub_ind_H_p = input("H[] -->")

reactante_1= ("C" + str(sub_ind_C_r) + "H" + str(sub_ind_H_r))
reactante_2=("O" + str(sub_ind_O_r))
producto_1=("C" + str(sub_ind_C_p) + "O" + str(sub_ind_O_p1))
producto_2=("O" + str(sub_ind_O_p2) + "H" + str(sub_ind_H_p))

ecuacion_con_indices= (reactante_1,"+",reactante_2,"-->",producto_1,"+",producto_2)
print("\nLa ecuación con los indices ingresados es: ", ecuacion_con_indices)

#equilibramos estequiometricamente con Chempy
#obtenemos los coeficientes
reactantes,productos=balance_stoichiometry({reactante_1, reactante_2} , {producto_1, producto_2})

coef1=reactantes[reactante_1]
coef2=reactantes[reactante_2]
coef3=productos[producto_1]
coef4=productos[producto_2]

print("\nEl coeficiente de ", reactante_1 ," es : ",coef1)
print("El coeficiente de ", reactante_2 ," es :" ,coef2)
print("El coeficiente de ", producto_1 ," es :",coef3)
print("El coeficiente de ", producto_2 ," es :",coef4)

molecularidad_reactantes = coef1 + coef2
molecularidad_productos =  coef3+ coef4

print("\nLa molecularidad de la reacción es : ",molecularidad_reactantes)
print("La molecularidad del producto es :" ,molecularidad_productos)

```

Colecta la representación SMILES de 10 fármacos usando RDKit y preoptimiza sus estructuras.
